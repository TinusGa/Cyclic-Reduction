```python
import numpy as np
import matplotlib.pyplot as plt 
from scipy.linalg import solve
```

```python
k = 4
r = 2
p = 2**r
n = 2**k
h = int(2**(k-r))
print("n: ", n, " p: ", p, " h: ", h)

def random_tridiagonal_matrix():
    A = np.zeros((n+1, n+1))
    for i in range(n+1):
        A[i, i] = np.random.randint(1,10)
        if i > 0:
            A[i, i - 1] = np.random.randint(1,10)
        if i < n - 1:
            A[i, i + 1] = np.random.randint(1,10)
    return A

def permutation_matrix():
    P = np.zeros((n+p,n+1))
    I_h = np.eye(h+1)
    j = 0
    k = 0
    for i in range(p):
        if i == 0:
            P[0:h+1,0:h+1] = I_h
            j = h+1
            k = h+1
        elif i == p-1:
            P[-1-h:, -1-h:] = I_h
        else:
            k -= 1
            P[j:j+h+1, k:k+h+1] = I_h
            j += h+1
            k += h+1
        
    return P


P = permutation_matrix()
plt.imshow(P, cmap='gray')
```

```python
A = random_tridiagonal_matrix()
plt.imshow(A, cmap='hot')
```

```python
print(A.shape)
print(P.shape)

plt.imshow(P@A@P.T,cmap='hot')
```

```python
n=4
A = random_tridiagonal_matrix()
f = np.random.randint(1,10, n+1)
x = solve(A,f)
print(A)
print(f)
print(x)
```
